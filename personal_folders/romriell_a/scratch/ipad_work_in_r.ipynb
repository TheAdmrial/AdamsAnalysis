```r
# loading in libraries
library(tidyverse)
library(arrow)
library(lubridate)


#reading in the data
df_par0 <- read_parquet('/content/tracker_0.parquet')
df_par1 <- read_parquet('/content/tracker_1.parquet')
df_par2 <- read_parquet('/content/tracker_2.parquet')
df_par3 <- read_parquet('/content/tracker_3.parquet')
df_par4 <- read_parquet('/content/tracker_4.parquet')
df_par5 <- read_parquet('/content/tracker_5.parquet')
df_par6 <- read_parquet('/content/tracker_6.parquet')
df_par7 <- read_parquet('/content/tracker_7.parquet')
df_par8 <- read_parquet('/content/tracker_8.parquet')
df_par9 <- read_parquet('/content/tracker_9.parquet')
df_par10 <- read_parquet('/content/tracker_10.parquet')
df_par11 <- read_parquet('/content/tracker_11.parquet')
# combining all the datasets we have
all_dat <- bind_rows(df_par0
          , df_par1
          , df_par2
          , df_par3
          , df_par4
          , df_par5
          , df_par6
          , df_par7
          , df_par8
          , df_par9
          , df_par10
          , df_par11)
# View(all_dat)

# data manipulation
all_get_dat <- all_dat %>%
  filter(request_method %in% 'GET') %>% 
  select(-server_name
         , -starts_with('ajax')
         , -request_time_float) %>% 
  group_by(user_id) %>% 
  arrange(user_id,request_time) %>%
  mutate(
    request_time = as_datetime(`request_time`)
    ,next_time = lead(request_time)
    ,request_date = as_date(request_time)
    ,mins_on_page = as.numeric((next_time - request_time)/60)
    ,active_5_mins = mins_on_page <=5
    ,active_10_mins = mins_on_page <=10
  ) %>% 
  ungroup() %>% 
  filter(!is.na(mins_on_page))
View(all_get_dat)
active_users_dat <- all_get_dat %>% 
  group_by(request_date, user_id) %>% 
  mutate(is_user_active_today_5 = case_when(
    sum(active_5_mins) > 0 ~ 1
    ,sum(active_5_mins) <= 0 ~ 0
  )
  ,is_user_active_today_10 = case_when(
    sum(active_10_mins) > 0 ~1
    ,sum(active_10_mins) <= 0 ~0
  )
) %>% 
  distinct(is_user_active_today_5, is_user_active_today_10) %>%
  ungroup() %>% 
  group_by(request_date) %>% 
  mutate(
    number_active_5 = sum(is_user_active_today_5)
    , number_active_10 = sum(is_user_active_today_10)
  ) %>% 
  distinct(number_active_5, number_active_10) %>% 
  View()
```
